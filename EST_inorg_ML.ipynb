```python
# Load the dataset - Single salt
import pandas as pd

df = pd.read_excel('Supplementary_data_final.xlsx', sheet_name='Single_salt')

# Load machine learning models - Single salt
import pickle
import shap

features = ['ionic_radius+', 'ionic_radius-', 'hyd_radius+', 'hyd_radius-','d_p-d_pi', 'charge_product', 'initial_conc', 'ionic_strength', 'MWCO', 'contact_angle', 'meas_T']

filename = 'xgb_single.sav'
ml_model = pickle.load(open(filename, 'rb'))

explainer = shap.TreeExplainer(ml_model)
shap_values = explainer.shap_values(df[features])

shap.summary_plot(shap_values, df[features], feature_names=df[features].columns, show=False)

shap_values = pd.DataFrame(shap_values)
shap_features = [feature+'-SHAP' for feature in features]
shap_values.columns = shap_features.copy()

shap_values = pd.concat([df[features], shap_values], axis=1)
```
